# **Load Dataset into Pandas**

```python
import pandas as pd

df = pd.read_csv('Cyber_Threats_Data.csv')
df.head()
```

# **Create SQLite Database & Table**

```python
import sqlite3

conn = sqlite3.connect('cyber_security.db')

df.to_sql(
    name='network_logs',
    con=conn,
    if_exists='replace',
    index=False
)

print("Table created successfully")
```

# **Verify Table Structure**

```python
pd.read_sql("PRAGMA table_info(network_logs);", conn)
```

# **Total Number of Records**

```python
pd.read_sql("SELECT COUNT(*) AS total_records FROM network_logs;", conn)
```

# **Total Number of Cyber Attacks**

```python
pd.read_sql("""
SELECT COUNT(*) AS total_attacks
FROM network_logs;
""", conn)
```

# **Attacks Per Year**

```python
pd.read_sql("""
SELECT Year, COUNT(*) AS attack_count
FROM network_logs
GROUP BY Year
ORDER BY Year;
""", conn)
```

# **Sample Records**

```python
pd.read_sql("""
SELECT *
FROM network_logs
LIMIT 10;
""", conn)
```

# **Most Common Types of Cyber Attacks**

```python
pd.read_sql("""
SELECT Type_of_Attack, COUNT(*) AS frequency
FROM network_logs
GROUP BY Type_of_Attack
ORDER BY frequency DESC;
""", conn)
```

# **Country-wise Attack Distribution**

```python
pd.read_sql("""
SELECT Location, COUNT(*) AS attack_count
FROM network_logs
GROUP BY Location
ORDER BY attack_count DESC;
""", conn)
```

# **Device Used for Attacks**

```python
pd.read_sql("""
SELECT Device_of_Attack, COUNT(*) AS device_count
FROM network_logs
GROUP BY Device_of_Attack
ORDER BY device_count DESC;
""", conn)
```

# **Security Mechanisms Involved**

```python
pd.read_sql("""
SELECT Security_Type, COUNT(*) AS security_count
FROM network_logs
GROUP BY Security_Type
ORDER BY security_count DESC;
""", conn)
```

# **Attack Type vs Location**

```python
pd.read_sql("""
SELECT Location, Type_of_Attack, COUNT(*) AS attack_count
FROM network_logs
GROUP BY Location, Type_of_Attack
ORDER BY attack_count DESC;
""", conn)
```

# **Most Targeted IP Addresses**

```python
pd.read_sql("""
SELECT Target_IP_Address, COUNT(*) AS times_targeted
FROM network_logs
GROUP BY Target_IP_Address
ORDER BY times_targeted DESC
LIMIT 10;
""", conn)
```

# **Repeated Attackers**

```python
pd.read_sql("""
SELECT Attacker_IP_Address, COUNT(*) AS attack_count
FROM network_logs
GROUP BY Attacker_IP_Address
ORDER BY attack_count DESC
LIMIT 10;
""", conn)
```

# **Year-wise Attack Trend**

```python
pd.read_sql("""
SELECT Year, COUNT(*) AS attacks
FROM network_logs
GROUP BY Year
ORDER BY Year;
""", conn)
```

# **Most Common Attack per Year**

```python
pd.read_sql("""
SELECT Year, Type_of_Attack, COUNT(*) AS count
FROM network_logs
GROUP BY Year, Type_of_Attack
ORDER BY Year, count DESC;
""", conn)
```

# **Countries with High Attack Activity**

```python
pd.read_sql("""
SELECT Location, COUNT(*) AS attack_count
FROM network_logs
GROUP BY Location
HAVING attack_count > 50
ORDER BY attack_count DESC;
""", conn)
```

# **Device vs Security Type Analysis**

```python
pd.read_sql("""
SELECT Device_of_Attack, Security_Type, COUNT(*) AS count
FROM network_logs
GROUP BY Device_of_Attack, Security_Type
ORDER BY count DESC;
""", conn)
```
